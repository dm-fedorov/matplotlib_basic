```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

x_coords = np.arange(10.5, 30.5, 0.5)
y_coords = x_coords

# Построить график.
plt.plot(x_coords, y_coords)

# Показать график.
plt.show()
```
