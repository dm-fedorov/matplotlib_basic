Найдите сумму всех значений функции y(x) = x^2 + 3 на интервале от 10 до 30 с шагом 2.2. 

Постройте график функции на указанном интервале. 

```python
import numpy as np

x = np.arange(10, 30, 2.2)
print(x)
y = np.power(x, 2) + 3
print(y)
print(np.sum(y))
```

```python
from numpy import pi
import math
import matplotlib.pyplot as plt
%matplotlib inline

x_coords = np.arange(10, 30, 2.2)
y_coords = np.power(x_coords, 2) + 3

# Построить график.
plt.plot(x_coords, y_coords)

# Показать график.
plt.show()
```
