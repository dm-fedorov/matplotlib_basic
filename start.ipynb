Текущие границы рисунка слишком жесткие, и мы хотим освободить место, чтобы четко видеть все точки данных.

```python
import numpy as np
import matplotlib.pyplot as plt

# plt.figure(figsize=(8, 6), dpi=80)
plt.figure(figsize=(10, 6), dpi=80)

# Create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

#plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")

#plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-")

# Set x limits
plt.xlim(X.min() * 1.1, X.max() * 1.1)

# Set x ticks
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))

# Set y limits
plt.ylim(C.min() * 1.1, C.max() * 1.1)

# Set y ticks
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

# Save figure using 72 dots per inch
# plt.savefig("exercise_2.png", dpi=72)

# Show result on screen
plt.show()
```

Текущие тики не идеальны, потому что они не показывают интересные значения (+/- π, + / - π / 2) для синуса и косинуса. Мы изменим их так, чтобы они показывали только эти значения.

```python
import numpy as np
import matplotlib.pyplot as plt

# plt.figure(figsize=(8, 6), dpi=80)
plt.figure(figsize=(10, 6), dpi=80)

# Create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

#plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")

#plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-")

# Set x limits
plt.xlim(X.min() * 1.1, X.max() * 1.1)

# Set x ticks
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])

# Set y limits
plt.ylim(C.min() * 1.1, C.max() * 1.1)

# Set y ticks
plt.yticks([-1, 0, +1])

# Save figure using 72 dots per inch
# plt.savefig("exercise_2.png", dpi=72)

# Show result on screen
plt.show()
```

Ticks теперь правильно размещены, но их ярлык не очень четкий. Можно предположить, что 3.142 - это π, но было бы лучше сделать это явным. Когда мы устанавливаем значения Ticks, мы также можем предоставить соответствующую метку во втором списке аргументов. Обратите внимание, что мы будем использовать latex, чтобы обеспечить хорошую визуализацию метки.

```python
import numpy as np
import matplotlib.pyplot as plt

# plt.figure(figsize=(8, 6), dpi=80)
plt.figure(figsize=(10, 6), dpi=80)

# Create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

#plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")

#plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-")

# Set x limits
plt.xlim(X.min() * 1.1, X.max() * 1.1)

# Set x ticks
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# Set y limits
plt.ylim(C.min() * 1.1, C.max() * 1.1)

# Set y ticks
plt.yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])

# Save figure using 72 dots per inch
# plt.savefig("exercise_2.png", dpi=72)

# Show result on screen
plt.show()
```

Spines - это линии, соединяющие отметки оси и отмечающие границы области данных. Их можно размещать в произвольных положениях и до сих пор они находились на границе оси. Мы изменим это, поскольку хотим, чтобы они были посередине. Поскольку их четыре (верхний / нижний / левый / правый), мы отбросим верхний и правый, установив для них цвет «Нет», и переместим нижний и левый в координату 0 в координатах пространства данных.

```python
import numpy as np
import matplotlib.pyplot as plt

# plt.figure(figsize=(8, 6), dpi=80)
plt.figure(figsize=(10, 6), dpi=80)

# Create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

#plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")

#plt.plot(X, S, color="green", linewidth=1.0, linestyle="-")
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-")

# Set x limits
plt.xlim(X.min() * 1.1, X.max() * 1.1)

# Set x ticks
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# Set y limits
plt.ylim(C.min() * 1.1, C.max() * 1.1)

# Set y ticks
plt.yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])

ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

# Save figure using 72 dots per inch
# plt.savefig("exercise_2.png", dpi=72)

# Show result on screen
plt.show()
```

Давайте добавим легенду в верхнем левом углу. Это требует только добавления ключевой метки аргумента (которая будет использоваться в поле легенды) к командам графика.

```python
import numpy as np
import matplotlib.pyplot as plt

# plt.figure(figsize=(8, 6), dpi=80)
plt.figure(figsize=(10, 6), dpi=80)

# Create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cosine")

plt.plot(X, S, color="red",  linewidth=2.5, linestyle="-", label="sine")

plt.legend(loc='upper left')

# Set x limits
plt.xlim(X.min() * 1.1, X.max() * 1.1)

# Set x ticks
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# Set y limits
plt.ylim(C.min() * 1.1, C.max() * 1.1)

# Set y ticks
plt.yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])

ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

# Save figure using 72 dots per inch
# plt.savefig("exercise_2.png", dpi=72)

# Show result on screen
plt.show()
```
